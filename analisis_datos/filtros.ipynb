```python
def convert_to_json_serializable(obj):
    if isinstance(obj, pd.Series):
        if obj.dtype == 'int64':
            return int(obj)
    # raise TypeError(f"Object of type {type(obj)} is not JSON serializable")
```

```python
def create_tree(df, features, level):
    if level == len(features):
        return None
    
    current_feature = features[level]
    unique_values = df[current_feature].unique()
    
    children = []
    for value in unique_values:
        child_df = df[df[current_feature] == value]
        child_tree = create_tree(child_df, features, level + 1)
        leaf_node = {
            "name": value,
            "children": [child_tree] if child_tree is not None else []
        }
        children.append(leaf_node)
    
    return {
        "name": current_feature,
        "children": children
    }
```

```python
feature_order = ['valence', 'key', 'tempo', 'loudness', 'explicit']

# Convert int64 columns to int
int_cols = df.select_dtypes(include=['int64']).columns
df[int_cols] = df[int_cols].astype(int)

# Convert int32 columns to int
int32_cols = df.select_dtypes(include=['int32']).columns
df[int32_cols] = df[int32_cols].astype(int)

# Crear el Ã¡rbol utilizando la funciÃ³n recursiva
tree = create_tree(df, feature_order, 0)

# Convertir el Ã¡rbol a formato JSON
json_tree = json.dumps(tree, default = convert_to_json_serializable)

# Imprimir el Ã¡rbol
print(json_tree)
```

```python
def create_tree(df, feature_order, index):
    if index == len(feature_order):
        return None

    feature = feature_order[index]
    children = []

    if feature == 'valence':
        groups = [
            {'name': 'valence <= 0.33'},
            {'name': '0.33 < valence < 0.66'},
            {'name': 'valence >= 0.66'}
        ]
        for group in groups:
            subset_df = df.copy()
            if group['name'] == 'valence <= 0.33':
                subset_df = subset_df[subset_df['valence'] <= 0.33]
            elif group['name'] == '0.33 < valence < 0.66':
                subset_df = subset_df[(subset_df['valence'] > 0.33) & (subset_df['valence'] < 0.66)]
            else:
                subset_df = subset_df[subset_df['valence'] >= 0.66]
            group['children'] = create_tree(subset_df, feature_order, index + 1)
            children.append(group)
    
    elif feature == 'key':
        for key_value in range(-1, 12):
            subset_df = df[df['key'] == key_value]
            group = {'name': f'key = {key_value}', 'children': create_tree(subset_df, feature_order, index + 1)}
            children.append(group)

    elif feature == 'tempo':
        groups = [
            {'name': 'tempo < 120 bpm'},
            {'name': 'tempo >= 120 bpm'}
        ]
        for group in groups:
            subset_df = df.copy()
            if group['name'] == 'tempo < 120 bpm':
                subset_df = subset_df[subset_df['tempo'] < 120]
            else:
                subset_df = subset_df[subset_df['tempo'] >= 120]
            group['children'] = create_tree(subset_df, feature_order, index + 1)
            children.append(group)

    elif feature == 'loudness':
        groups = [
            {'name': 'loudness < -10 dB'},
            {'name': 'loudness >= -10 dB'}
        ]
        for group in groups:
            subset_df = df.copy()
            if group['name'] == 'loudness < -10 dB':
                subset_df = subset_df[subset_df['loudness'] < -10]
            else:
                subset_df = subset_df[subset_df['loudness'] >= -10]
            group['children'] = create_tree(subset_df, feature_order, index + 1)
            children.append(group)

    elif feature == 'explicit':
        groups = [
            {'name': 'explicit = True'},
            {'name': 'explicit = False'}
        ]
        for group in groups:
            subset_df = df.copy()
            if group['name'] == 'explicit = True':
                subset_df = subset_df[subset_df['explicit'] == True]
            else:
                subset_df = subset_df[subset_df['explicit'] == False]
            group['children'] = create_tree(subset_df, feature_order, index + 1)
            children.append(group)

    return children
```

```python
# Definir el orden de las caracterÃ­sticas
feature_order = ['valence', 'key', 'tempo', 'loudness', 'explicit']

# Crear la estructura del Ã¡rbol
tree = {
    "name": "valence",
    "children": create_tree(df1, feature_order, 0)
}

# Imprimir el Ã¡rbol
import json
json_tree = json.dumps(tree, default = convert_to_json_serializable)
print(json_tree)
```

```python
output_file = "tree.json"

# Write the JSON tree to the file
with open(output_file, "w") as file:
    json.dump(json_tree, file)

print(f"JSON tree exported to {output_file} successfully.")
```

```python
import csv
import json

# Read the CSV file and parse the data
data = []
with open('Afterlife.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        data.append(row)

def build_nested_structure(song):
    # Create the root node for each valence category
    root_node = {
        'name': '',
        'children': []
    }

    # Categorize songs based on valence value
    if float(song['valence']) <= 0.33:
        root_node['name'] = 'valence <= 0.33'
    elif float(song['valence']) >= 0.66:
        root_node['name'] = 'valence >= 0.66'
    else:
        root_node['name'] = '0.33 < valence < 0.66'

    # Create child nodes for each key value
    for key in range(-1, 12):
        key_node = {
            'name': 'Key ' + str(key),
            'children': []
        }
        root_node['children'].append(key_node)

        # Create child nodes for tempo groups
        tempo_node1 = {
            'name': 'Tempo <120',
            'children': []
        }
        key_node['children'].append(tempo_node1)

        tempo_node2 = {
            'name': 'Tempo >120',
            'children': []
        }
        key_node['children'].append(tempo_node2)

        # Create child nodes for loudness groups
        loudness_node1 = {
            'name': 'Loudness <-10',
            'children': []
        }
        tempo_node2['children'].append(loudness_node1)

        loudness_node2 = {
            'name': 'Loudness >-10',
            'children': []
        }
        tempo_node2['children'].append(loudness_node2)

        # Create child nodes for explicit groups
        explicit_node1 = {
            'name': 'Explicit',
            'children': []
        }
        loudness_node2['children'].append(explicit_node1)

        explicit_node2 = {
            'name': 'No Explicit',
            'children': []
        }
        loudness_node2['children'].append(explicit_node2)

        # Add the song node to the appropriate explicit group
        song_node = {
            'name': song['name'],
            'artist': song['artist'],
            'size': int(song['popularity'])*10,
        }
        if song['explicit'].lower() == 'true':
            explicit_node1['children'].append(song_node)
        else:
            explicit_node2['children'].append(song_node)

    return root_node

# Build the nested structure for each song in the data list
nested_structures = []
for song in data:
    nested_structure = build_nested_structure(song)
    nested_structures.append(nested_structure)

# Convert the nested structures to JSON
json_output = json.dumps(nested_structures, indent=4)

# Print the JSON output
print(json_output)
```
